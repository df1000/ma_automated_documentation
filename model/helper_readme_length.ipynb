```python
import json
import pandas as pd
import numpy as np
```

```python
def open_json(path):
    '''
    Function which opens and read a JSON file.

    Args:
        path: Path to JSON file.

    Return:
        loaded_data or {}
    '''
    try: # try to open file
        with open(path, 'r') as file: # open and load JSON file
            loaded_data = json.load(file) 
        return loaded_data # save loaded content in variable loaded_data
    except json.decoder.JSONDecodeError: # raise exception if JSONDecodeError --> file is empty
        return {}
```

```python
loaded_data = open_json(path='../data/df_repos_counts_filtered.json') # load json with filtered repository
df = pd.DataFrame(loaded_data) # create Pandas dataframe with loaded_data
```

```python
df = df.drop(['source_code_comments', 'source_code', 'source_code_cleaned_comments', 'source_code_cleaned', 'comments', 'formatting', 'sum'], axis=1)
```

```python
cnt = 0
new_rows = []
for _, row in df.iterrows():
    # if cnt ==2:
    #     break

    repo_name = row['repo_name']
    repo_owner = row['repo_owner']

    # evaluation for generated readme
    readme_generated_data = open_json(path=f'../data/output_readme_data/{repo_owner}_{repo_name}_output.json') # call open_json()
    readme_generated = readme_generated_data['readme']

    # evaluation for original readme
    readme_original_data = open_json(path=f'../data/input_readme_data/{repo_owner}_{repo_name}.json')
    readme_original = readme_original_data['readme']

    new_row = {
        'repo_name': repo_name,
        'repo_owner': repo_owner,
        'repo_owner_generated_test': readme_generated_data['repo_owner'],
        'repo_name_generated_test': readme_generated_data['repo_name'],
        'readme_generated': readme_generated,
        'readme_generated_cnt': len(readme_generated),
        'repo_owner_original_test': readme_original_data['repo_owner'],
        'repo_name_original_test': readme_original_data['repo_name'],
        'readme_original': readme_original,
        'readme_original_cnt': len(readme_original) if readme_original is not None else np.nan

    }

    new_rows.append(new_row)

    #cnt +=1

new_df = pd.DataFrame(new_rows)   

   
```
